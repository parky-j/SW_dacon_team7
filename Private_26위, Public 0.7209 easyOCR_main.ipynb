### Train-Val Split

```python
import shutil
import os
import natsort

# dir1 -> dir2 로 파일 Move
dir1 = "/content/drive/MyDrive/data/train/"
dir2 = "/content/drive/MyDrive/김영서/Dataset_da_hub/train_raw/train/"

files_list = os.listdir(dir1) # dir1의 파일 목록을 list로 받는다.
after_order_list = natsort.natsorted(files_list) # asc 정렬
files = after_order_list[4340:]

for file in files:
    shutil.copy(dir1 + file, dir2 + file) # 하나씩 순서대로 복사
    print ('Creating...' + dir2 + file)
```
